```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time

# Mendeklarasikan list kosong untuk masing-masing kolom
product_names = []
product_prices = []
product_categories = []
product_colors = []

# Mendeklarasikan chrome sebagai browser yang akan digunakan
driver = webdriver.Chrome()

# Mendeklarasikan url untuk masing-masing page 
url = f"https://www.nike.com/id/w?q=shoe&vst=shoe"
    
# Menjalankan chrome dengan link secara otomatis
driver.get(url)
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
```

```python
# Menyimpan semua HTML element di variabel product_container
product_container = soup.find_all("div",{"data-testid":"product-card"})
# time.sleep(3)
# driver.execute_script('window.scrollBy(0, 100000)')
# time.sleep(5)


for container in product_container:
    
    # Mengambil data nama produk
    if container.find_all("div",{"class":"product-card__title"}) != None:
        product_names += [name.text for name in soup.find_all("div",{"class":"product-card__title"})] #product_names.append(container.find("div",{"class":"product-card__title"}).text)
    else:
        product_names += (None)

    # Mengambil data harga produk
    if container.find_all("div",{"data-testid":"product-price"}) != None:
        product_prices += [price.text for price in soup.find_all("div",{"data-testid":"product-price"})]
    elif container.find_all("div",{"data-testid":"product-price"}) != None:
        product_prices += [price.text for price in soup.find_all("div",{"data-testid":"product-price"})]
    else:
        product_prices += (None)
        
    # Mengambil data kategori produk
    if container.find_all("div",{"class":"product-card__subtitle"}) != None:
        product_categories += [category.text for category in soup.find_all("div",{"class":"product-card__subtitle"})]
    else:
        product_categories += (None)
        
    # Mengambil data jumlah warna produk
    if container.find_all("div",{"class":"product-card__product-count"}) != None:
        product_colors += [color.text for color in soup.find_all("div",{"class":"product-card__product-count"})]
    else:
        product_colors += (None)
```

```python
# Me-limit semua data menjadi 100 row
max_length = 100
product_names = product_names[0:max_length]
product_prices = product_prices[0:max_length]
product_categories = product_categories[0:max_length]
product_colors = product_colors[0:max_length]
```

```python
# Membuat dataframe dari masing-masing data yang sudah diambil
df = pd.DataFrame({"product_name":product_names,
                        "product_price":product_prices,
                        "product_category":product_categories,
                        "product_color":product_colors})

# Mendisplay sebagian data awal dar dataframe yang sudah dibuat
df.head(27)

        
# df_test
```

```python
# Mengambil angka total terjual dari HTML elemen yang menyimpan total barang terjual
new_price_column = df["product_price"].str.replace("Rp\xa0", "")
new_price_column = new_price_column.str.replace(".", "")
new_price_column = new_price_column.astype("int64")

# Menggabungkan kolom-kolom yang dibuat sebelumnya ke dataframe utama
df["product_price"] = new_price_column
```

```python
# Mengimport data ke CSV
df.to_csv('shoe_data.csv', index=False)
```
